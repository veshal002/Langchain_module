AI agents normally store the conversation in something called STATE (memory). But without state it is a agent with amnesia.
To provide state there is something called CHECKPOINTER that is given to the bot as it acts as a checkpoint thus retaining information and continuing the conversation.

```python
from langgraph.checkpoint.memory import InMemorySaver
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain.agents import create_agent
from dotenv import load_dotenv # To load variables inside the .env file
from pprint import pprint # To format messy dictionary outputs to human readable output

load_dotenv()

model = ChatGoogleGenerativeAI(
    model="gemini-2.5-flash-lite", # LLM the brain of the agent
    timeout=10
)

agent = create_agent(
    model=model,
    checkpointer=InMemorySaver() # Agent with memory
)
```

```python
from langchain.messages import HumanMessage

question = HumanMessage(content="I am veshal and i love bikes but i love cars more.")
config={"configurable":{"thread_id":"1"}}

response=agent.invoke(
    {"messages":[question]},
    config
)
```

```python
pprint(response)
```

```python
question = HumanMessage(content="Can you tell me what my name is and what i love?")


response_2=agent.invoke(
    {"messages":[question]},
    config
)
```
