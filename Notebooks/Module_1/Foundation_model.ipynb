```python
from langchain.chat_models import init_chat_model
from langchain_google_genai import ChatGoogleGenerativeAI
from dotenv import load_dotenv # To load variables inside the .env file
from pprint import pprint # To format messy dictionary outputs to human readable output

load_dotenv()

model = ChatGoogleGenerativeAI(
    model="gemini-2.5-flash-lite",
    temperature=0.5,
    timeout=10
)

response = model.invoke("What's the capital of the Moon?")

print(response.content)
print(response.response_metadata)
print(response.usage_metadata)
```

```python
from langchain.agents import create_agent

agent=create_agent(model=model)
```

```python
from langchain.messages import HumanMessage

response = agent.invoke(
    {"messages": [HumanMessage(content="What's the capital of the Moon?")]}
)
```

```python
from pprint import pprint

pprint(response)
```

```python
print(response['messages'][-1].content)
```

```python
# Stream that prints out output as it recieves instead of collecting fully and printing it

for token,metadata in agent.stream(
    {"messages": [HumanMessage(content="What's the fun fact abou the moon?")]},
    stream_mode='messages'
):

    # token is a message chunk with token content
    # metadata contains which node produced the token
    if token.content: # check if there's actual content
            print(token.content,end="",flush=True) # print token
```

TEMPERATURE - More temperature more creative, lower ones make it deterministic
MAXTOKENS - Maximum tokens allowed thus limiting the lenght of the response
TIMEOUT - Maximum time to wait before terminating the waiting of response
MAX RETRIES - Maximum number of times to retry.


```python
from langchain.agents import create_agent
from langchain.messages import HumanMessage
from pydantic import BaseModel

class CapitalInfo(BaseModel):
    name: str
    location: str
    vibe: str
    economy: str
agent = create_agent(
    model=model,
    system_prompt="You are a Creative science fiction writer, Create a capital city at users request.",
    response_format=CapitalInfo # format at which the output should be in 
)
question = HumanMessage(content="What is the capital of the Moon?")

response =agent.invoke(
    {"messages":[question]}
)

response["structured_response"]
```

```python
capital_info= response["structured_response"]
capital_city = capital_info.name
capital_city_location = capital_info.location

print(f"The city called {capital_city} is located in {capital_city_location}")
```
